```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector, visualize_transition
from numpy import pi
```

```python
# function to show gate circuit & its state in bloch sphere
def show_gate(qc):
    display(qc.draw('mpl'))
    display(plot_bloch_multivector(qc))
```

```python
# I dentity Gate
qc = QuantumCircuit(1)
qc.i(0)

show_gate(qc)
```

```python
# Pauli X Gate | NOT Gate
qc = QuantumCircuit(1)
qc.x(0)

show_gate(qc)
```

```python
# Pauli Y Gate
qc = QuantumCircuit(1)
qc.y(0)

show_gate(qc)
```

```python
# Pauli Z Gate
qc = QuantumCircuit(1)
qc.z(0)

show_gate(qc)
```

```python
# Hadamard H Gate
qc = QuantumCircuit(1)
qc.h(0)

show_gate(qc)
```

```python
# Rotation X Gate
qc = QuantumCircuit(1)
qc.rx(pi, 0)

show_gate(qc)
```

```python
# Rotation Y Gate
qc = QuantumCircuit(1)
qc.ry(pi, 0)

show_gate(qc)
```

```python
# Rotation Z Gate
qc = QuantumCircuit(1)
qc.rz(pi, 0)

show_gate(qc)
```

```python
# CNOT | Controlled Not Gate
qc = QuantumCircuit(2)
qc.cx(0, 1)

show_gate(qc)
```
